```python
X = df.drop('Salary',axis=1)
Y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(len(X),len(x_train),len(x_test))
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
df1 = pd.DataFrame()
df1['Actual'] = y_test
df1['Predicted'] = y_pred
df1['Error'] = df1.Actual - df1.Predicted
```

```python
df1
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print('Mean Squared Error: ',mean_squared_error(y_test,y_pred),'\nMean Absolute Error: ',mean_absolute_error(y_test,y_pred),'\nR2 Score: ',r2_score(y_test,y_pred),'\nRoot Mean Square Error (RMSE): ',mean_squared_error(y_test,y_pred)**0.5)
```

```python
r2 = r2_score(y_test,y_pred)
```
