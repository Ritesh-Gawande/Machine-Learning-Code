Modeling

```python
x = df.drop('Outcome',axis=1)
y= df.Outcome
```

```python
y.value_counts()
```

```python
x.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train_raw,x_test_raw,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from re import X
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scal = scaler.fit_transform(x_train_raw)
X_test = scaler.transform(x_test_raw)
```

```python
x_scal
```

```python
x_train = pd.DataFrame(data=x_scal,columns=x_train_raw.columns)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_test = pd.DataFrame(data=X_test,columns=x_test_raw.columns)
y_pred = model.predict(x_test)
```

```python
X1 = pd.DataFrame()
X1[0] = y_pred
X1[1] = y_test.values
X1[2] = y_test.values - y_pred
```

```python
X1
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
```
